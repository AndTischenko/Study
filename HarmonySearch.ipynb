```python
import numpy as np
import random
def func (x):
    return 4*pow((x[0]-5),2)+pow((x[1]-6),2)
hms = 20 #количество гармоник
hmcr = 0.5 # вероятность выбора из гармоник в памяти от 0 до 1
par = 0.5 # вероятность модификации от 0 до 1
bw = 2 #величина модификации
eps = 0.01 #достигаемая точность
hm =  [np.random.uniform(-10, 10, 2) for i in range(hms)]
hm_y = [func(hm[i]) for i in range(hms)] 
k = 0 # cчетчик итераций
print("Начальные данные:\n")
for i in range(hms):
    print("Гармоника:{}\nЗначение функции: {}\n".format(hm[i],hm_y[i]))
while(min(hm_y)>eps):
    print("Итерация: {}".format(k))
    max_index = np.argmax(hm_y)
    h_new = [0 for i in range(len(hm[0]))]
    for i in range(len(h_new)):
        e = random.random()
        if(e < hmcr):
            h_new[i] = hm[random.randint(0,len(hm)-1)][i]
            e1 = random.random()
            if(e1 < par):
                h_new[i] = h_new[i]*random.uniform(-1,1) * bw
        else:
            h_new[i] = random.random()
    print("Новая гармоника: {}".format(h_new))
    print("{}<{}?".format(func(h_new),max(hm_y)))
    if(func(h_new) < max(hm_y)):
        hm[max_index] = h_new
        hm_y =[func(hm[i]) for i in range(hms)] 
        print("\nИзмененные данные:")
        for i in range(hms):
            print("Гармоника:{}\nЗначение функции: {}\n".format(hm[i],hm_y[i]))
    else:
        print("\nДанные не были изменены:")
        for i in range(hms):
            print("Гармоника:{}\nЗначение функции: {}\n".format(hm[i],hm_y[i]))
    k+=1
min_index = np.argmin(hm_y)
print("Минимум:\nХ1={}\nX2={}\nY={}\n".format(hm[min_index][0],hm[min_index][1],min(hm_y)))
```
