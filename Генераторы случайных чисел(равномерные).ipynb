```python
import random
import numpy as np
import matplotlib.pyplot as plt
import numpy as np
NN=1000 
pred_x = random.randint(1000000000,99999999999)
def randgen(x):
    a = 843314861
    c = 4513816693
    return float(((a*x + c)%10)*0.1)
```

```python
X= []
X.append(randgen(pred_x))
for i in range(1,NN):# Формирование выборки 
    X.append(randgen(X[i-1]))
    print(X[i])
X1= np.random.uniform(0,1,NN)# Выборки сгенерированных станд.функцией Python
#сортировка копий массивов для графиков
X1_sort = [] 
X_sort = []
for i in range(0,NN): 
    X_sort.append(X[i])
    X1_sort.append(X1[i])
X_sort.sort()
X1_sort.sort()
```

### Проверка апериодичности 

```python
X2= []
X2.append(randgen(X[0]))
for i in range(1,NN):
    X2.append(randgen(X2[i-1]))
for i in range(0,NN):
    if(X[i]==X2[i]):
        print(i)
```

### Подсчет интервалов для проверки равномерности плотности вероятностей

```python
import math
math.pi
K = math.floor(1+3.3*math.log10(float(NN))) 
print(K)
```

### Подсчет числа значений в каждом интервале Ni

```python
N=[]
N_st=[]
for i in range (0,K):
    N.append(0)
    N_st.append(0)
for i in range(0,NN):
    ri = math.floor(X[i]*K)
    ri_st = math.floor(X1[i]*K)
    N[ri]=N[ri] + 1
    N_st[ri_st]=N_st[ri_st] + 1
    ri = 0
    ri_st = 0
print(N)
print(N_st)
```

### Xi-квадрат критерий

```python
Xi = 0
Xi_st = 0
for i in range (K):
    Xi = Xi + (((N[i]-NN/K)**2)/(NN/K))
    Xi_st = Xi_st + ((N_st[i]-NN/K)**2/(NN/K))
print("Хи-квадрат критерий сгенерированный выборки")
print(Xi)
print("Хи-квадрат критерий выборки сегенерированной стандартной библиотекой Python")
print(Xi_st)
```

```python
tmp = np.array(X)
print("Числовые характеристики сгенерированной выборки")
print("Дисперсия:",np.var(tmp))
print("Мат.ожидание",np.mean(tmp))
print()
print("Числовые характеристики выборки созданной стандартной библиотекой Python")
print("Дисперсия:",np.var(X1))
print("Мат.ожидание",np.mean(X1))
```
